```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
driver = webdriver.Chrome()

#create a url
url = r"https://www.nba.com/stats/players/advanced/?sort=GP&dir=-1"

#get the url
driver.get(url)


#select variable and find the xpath of a listBox of pages 4xmple


select = Select(driver.find_element_by_xpath(r"/html/body/main/div/div/div[2]/div/div/nba-stat-table/div[1]/div/div/select"))



select.select_by_index (0)


#check and choose the class, find table
src = driver.page_source
parser = BeautifulSoup(src, "lxml")
table = parser.find("div", attrs = {"class" : "nba-stat-table__overflow"} )

#to get the textValues
headers = table.findAll('th')

#not get the first header, which is hidden value
headersList = [header.text.strip() for header in headers[1:]]

#new list without hidden values
headersList_notRanked = []
for header in headersList :
    if not 'RANK' in header :
        headersList_notRanked.append(header)


headersList_notRanked

```

```python
#rows table

rows = table.findAll('tr')[1:]


```

```python
# list of players stats
players_stats = [[td.getText().strip() for td in rows[i].findAll ('td')[1:]]for i in range (len(rows))]
```

```python
#erase hidden values in the dataframe


headersList_notRanked = headersList_notRanked [:-5]
```

```python
#clean data and set dataframe

stats = pd.DataFrame(players_stats, columns = headersList_notRanked)
```
